```python
data_country=pd.read_excel("/content/Country-Code.xlsx")
data_country.head()
```

```python
final=pd.merge(data,data_country,on="Country Code",how="left")
final.head()
```

```python
final.columns
```

```python
final.dtypes
```

```python
final.shape
```

```python
country_names=final.Country.value_counts().index
```

```python
country_values=final.Country.value_counts().values
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

OBSERVATIONS:::- Zomato maximum records are from India Then USA AND UK.

```python
final.columns
```

```python
final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_rating=final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating count"})
```

```python
final_rating.head()
```

```python
import matplotlib
matplotlib.rcParams["figure.figsize"]=(12,6)
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",data=final_rating,hue="Rating color",palette=["white","red","orange","yellow","green","green"])
```

Observation::-
1. Not rated are maximum
2. Maximum number of rating are between 2.5 and 3.9

```python
sns.countplot(x="Rating color",data=final_rating,palette=["white","red","orange","yellow","green","green"])
```
